### Relax Inc Data Science Challenge

__Taks__ : Defining an "adopted user" as a user who has logged into the product on three separate days in at least one sevenÂ­day period, identify which factors predict future user adoption.

```python
import pandas as pd
import numpy as np
```

```python
users=pd.read_csv("users.csv")
engagement=pd.read_csv("takehome_user_engagement.csv")
```

```python
users.head()
```

```python
engagement.head()
```

```python
engagement["time_stamp"]=pd.to_datetime(engagement["time_stamp"])
```

```python
engagement["date"]=engagement["time_stamp"].dt.date
```

### 1/ Determine which users have been adopted

```python
adopted=[]
for user_id in engagement["user_id"].unique() : 
    if len(engagement[engagement["user_id"]==user_id]) < 3 :
        adopted.append(0)
    else : 
        
        logs=engagement[engagement["user_id"]==user_id]["date"]
        logs=(logs.shift(-2) - logs).dt.days
        if ((logs<8).sum()) > 1 :
            adopted.append(1)
        else : 
            adopted.append(0)
        
        
        
    
```

```python
adopted=pd.DataFrame({"object_id": engagement["user_id"].unique(), "adopted": adopted})
```

```python
adopted.head()
```

```python
adopted
```

```python
users=pd.merge(users,adopted,how="right",on="object_id")
```
